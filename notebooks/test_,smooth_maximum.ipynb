```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

def max_function(x):
    return np.maximum(x, 0)

def log_sum_exp(x):
    return np.log(np.exp(x) + np.exp(0))

def smooth_relu(x, alpha=0.001):
    return np.log(1 + np.exp(x / alpha)) * alpha

def smooth_max(x, alpha=0.5):
    return alpha * x + (1 - alpha) * 0

X = np.linspace(-0.01, 0.01, 400)

Z_max = max_function(X)
Z_log_sum_exp = log_sum_exp(X)
z_smooth_relu = smooth_relu(X)

df = pd.DataFrame(
    zip(Z_max, Z_log_sum_exp,z_smooth_relu, X),
    columns = ["max", "z_log", "z_smooth_relu", "x"]
)

plt.figure(figsize=(10,5))

sns.lineplot(
    x = "x",
    y = "max",
    data = df
)
sns.lineplot(
    x = "x",
    y = "z_smooth_relu",
    data = df
)
plt.show()
```
